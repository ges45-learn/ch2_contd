````python
#q.2.4.3 ranking a list
'''
The core trick relies on the behavior of the list method scores.index(score)
The Rule of .index(): It searches the list from left to right and returns the index of the first occurrence of the value it finds.
If a value appears multiple times, .index() will always return the index of its first appearance.

| Iteration | `score` | Position in List | `scores.index(score)` | `+ 1` (Rank) | Explanation |
| --- | --- | --- | --- | --- | --- |
| **1** | **87** | Index 0 | `0` | **1** | First time seeing 87. Index is 0. |
| **2** | **75** | Index 1 | `1` | **2** | First time seeing 75. Index is 1. |
| **3** | **75** | Index 2 | `1` | **2** | **Tie!** `.index()` looks back and finds the *first* 75 at index 1. |
| **4** | **50** | Index 3 | `3` | **4** | First time seeing 50. Index is 3. (Rank 3 is skipped). |
| **5** | **32** | Index 4 | `4` | **5** | First time seeing 32. Index is 4. |
| **6** | **32** | Index 5 | `4` | **5** | **Tie!** `.index()` looks back and finds the *first* 32 at index 4. |


'''

scores = [87, 75, 50, 32, 32]
ranks = []
for score in scores:
    # Find the position of the score and calculate its rank:
    #    - scores.index(score) looks up the index of the FIRST occurrence of that score.
    #    - We add + 1 because Python indices start at 0, but ranks start at 1.
    calculated_rank = scores.index(score) + 1
    ranks.append(calculated_rank) # add ranks to list
print(ranks)

````

````python
#q2.4.4 madhava-leibniz series
#initialisation
pi = 0
pi2 = 0

for k in range(20): # loop 20 times (from k = 0 to k = 19), each iteration calculates one term of the infinite series
  #calculate the kth term and add it to the running total
  # pow(-3, -k) represents (-3)^(-k) which handles the alternating sign and shrinking value
  # (2 * k + 1) generates the odd denominators
  pi += pow(-3,-k) / (2 * k + 1)
  pi2 +=  (-1)**k / (3**k * (2 * k + 1))

pi *= m.sqrt(12) # multiply sum by sqrt(12) to complete formula
pi2 *= m.sqrt(12)

print("pi = ", pi2)

````

````python
"""

This code pairs up two separate lists (months and hours of sunshine), sorts them from the sunniest month to the least sunny month, and prints the results in a formatted layout.

Here is how the core trick works under the hood before we look at the annotations:

* **`zip(sun, months)`**: This pairs each sunshine number with its corresponding month name into a single object, like `(44.7, 'Jan')`. We put `sun` first because Python sorts pairs based on their **first element**.
* **`sorted(..., reverse=True)`**: This sorts the pairs in descending order (highest sunshine hours to lowest).
* **`'{}: {:5.1f} hrs'`**: This is a string formatting tool. `{:5.1f}` means "format this float to take up at least 5 spaces total, with exactly 1 digit after the decimal point." This keeps the numbers neatly aligned.

---

"""

### Annotated Code

#python
# 1. Define a list of abbreviations for all 12 months
months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]

# 2. Define a list of average monthly sunshine hours matching the order of the months
sun = [
    44.7,
    65.4,
    101.7,
    148.3,
    170.9,
    171.4,
    176.7,
    186.1,
    133.9,
    105.4,
    59.6,
    45.8,
]

# 3. Create a loop that processes the sorted data pairs:
#    - zip(sun, months) merges the lists into pairs: (44.7, 'Jan'), (65.4, 'Feb'), etc.
#    - sorted(...) sorts these pairs based on the first item in the pair (the sunshine hours).
#    - reverse=True ensures the sorting goes from highest value to lowest value.
#    - 'for s, m' unpacks each sorted pair back into 's' (sunshine) and 'm' (month) variables.
for s, m in sorted(zip(sun, months), reverse=True):

    # 4. Print the month and its sunshine hours using strict formatting:
    #    - {}: prints the month string (m)
    #    - {:5.1f}: prints the float (s) right-aligned with 1 decimal place, taking up 5 characters of width
    #    - ' hrs': prints a literal suffix string after the number
    print("{}: {:5.1f} hrs".format(m, s))
````

````python
#2.6.1 redwood exemplar

!rm -rf ch2_contd # discards any folder named 'ch2_contd' to ensure a clean state
!git clone https://github.com/ges45-learn/ch2_contd.git # clones repo into local environment

data_file = 'ch2_contd/redwood-data.txt' # defines the path to the txt file containing the redwood data
#indexes of the fields in each record
NAME, LOCATION, DIAMETER, HEIGHT = 0,1,2,3 # great practice

f = open(data_file) # opens for reading
# skip the header lines
f.readline()
f.readline()

tree_data = [] # initialisation, a list that will store all parsed tree records
max_height, max_diameter = 0,0  # tracking variablse to keep score of the largest values found so far

#enumerate tracks the line number
for item, line in enumerate(f.readlines()): # loops through every remaining line
  records = line.split('\t') # splits the line by tabs into a list of strings
  diameter = records[DIAMETER] = float(records[DIAMETER]) # converts the diameter and height into decimal numbers
  height = records[HEIGHT] = float(records[HEIGHT])

  if diameter > max_diameter:
    max_diameter, max_diameter_index = diameter, item # max_diameter_index ensures the code remembers how to loop them up later
  if height > max_height:
    max_height, max_height_index = height, item  # max_height_index ensures the code remembers how to loop them up later
  tree_data.append(records)

highest_tree = tree_data[max_height_index]
print('the highest tree: {}, {} m in {}'.format(highest_tree[NAME], highest_tree[LOCATION], highest_tree[HEIGHT]))

widest_tree = tree_data[max_diameter_index]
print('the widest tree: {}, {} m in {}'.format(widest_tree[NAME],widest_tree[LOCATION], widest_tree[DIAMETER]))

f.close()
````
