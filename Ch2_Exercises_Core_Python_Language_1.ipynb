```python
#2.5.1 exemplar, pascal's triangle using a nested list

# INITIALISTION
nmax = 4 # 5 rows
# Initialize with the first two rows
rows = [[1], [1, 1]] # pre-loaded with first rows

# OUTERLOOP (ROW BY ROW)
for n in range(2, nmax + 1):
    rows.append([1]) # new row by placing 1 at the beginning

    prev_row = rows[n - 1] # gets the completed row right above so we can look at the numbers

# INNER LOOP (MATH & COLUMN CREATION), WHERE THE MAGIC HAPPENS

    for k in range(len(prev_row) - 1):
        rows[n].append(prev_row[k] + prev_row[k + 1])

    rows[n].append(1) # End the row with a 1

#'Pretty-print the rows as asterisks for odd numbers or else spaces'


for row in rows:
    s_row = []
    for x in row:
        if x % 2: # if number is odd x % 2 evaluates to --> True, so it appends *
            s_row.append('*')
        else:
            s_row.append(' ') # if even --> False
    # Optional: Center the output slightly for a better visual triangle
    print(''.join(s_row))
```

```python
#p2.5.5 determine tomorrow's date

# text extraction phase

today = '22/05/2024'

us_style_today = 0
uk_style_today = 1

# Split once up front to keep the code clean
date_parts = today.split('/')

if us_style_today == True:
    day = int(date_parts[1])
    month = int(date_parts[0])
    year = int(date_parts[2])
else:
    day = int(date_parts[0])
    month = int(date_parts[1])
    year = int(date_parts[2])

group_1 = [1, 3, 5, 7, 8, 10, 12]
group_2 = [4, 6, 9, 11]
group_3 = [2]

# CALCULATE LEAP YEAR UP FRONT: Moved here so it doesn't break the if/elif chain below
is_leap_year = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Main date rollover logic
if month in group_1:
    if day == 31:
        if month == 12:  # Safe catch for New Year's Eve
            month = 1
            year += 1
        else:
            month += 1
        day = 1
    else:
        day += 1

elif month in group_2:
    if day == 30:
        month += 1
        day = 1
    else:
        day += 1

elif month in group_3:  # Now this safely connects back to the main chain
    # Condition A: It's Leap Day (Feb 29) OR it's Feb 28 in a normal year
    if (is_leap_year and day == 29) or (not is_leap_year and day == 28):
        month = 3
        day = 1

    # Condition B: It's Feb 28 in a leap year -> Advance to Leap Day (Feb 29)
    elif is_leap_year and day == 28:
        day += 1

    # Condition C: Any other day in February (e.g., Feb 1st to 27th) -> Just add 1
    else:
        day += 1

print(f"Tomorrow is: {day:02d}/{month:02d}/{year}")

```

```python
def elec_config(Z=6):
  """
  Write a program to predict the electronic configurations of the elements up to
  Rutherfordium (N=104). The output for Titanium should be

  Ti: 1s2.2s2.2p6.3s2.3p6.4s2.3d2

  This function returns the electronic configuration of an atom with atomic
  number Z.
  Z is valid until Z = 104.

  Input:
  ------
  Z: int


  Output:
  ------
  config: str
  """
  # import element symbols
  import element_symbols as es
  symbols = es.element_symbols

  # Find the symbol associated with number Z
  atom_name = symbols[Z-1]
  elec_config = ""

  allocated = 0

  while allocated < Z:
    if allocated < 2:
      if elec_config[:2] != '1s':
        elec_config += '1s'
        allocated += 1
      else:
        elec_config += '2'
        allocated += 1


  config = atom_name + ": " + elec_config
  return config
```

```python
"""
use date to calculate the ESI of each bodies. which has the closet?

Input
-----
txt

Output
--------
str
"""

#setup the constants
earth_values = [1.0, 1.0, 1.0, 288]
earth_weights = [0.57, 1.07, 0.7, 5.58]


!rm -rf ch2_contd
!git clone https://github.com/ges45-learn/ch2_contd.git

data_file = 'ch2_contd/ex2-6-g-esi-data.txt'
f = open(data_file)
f.readline()
f.readline()
f.readline()

#indices
NAME, MASS, RADIUS, DENSITY, GRAVITY, V_ESC, A, TSURF, TEQ = range(9)

#initialisation of list that will store all parsed tree records
esi = 0 # tracking to keep score
closest_planet = ""


#enumerate tracks the line number
for item, line in enumerate(f.readlines()): # loops through remaining line
  records = line.strip().split('\t') # splits line by tabs into list of strings

  if len(records) < 9:
    continue

  name = records[NAME]
  radius = float(records[RADIUS])
  density = float(records[DENSITY])
  v_esc = float(records[V_ESC])
  tsurf = float(records[TSURF])

  planet = [radius, density, v_esc, tsurf]

  tracker_product = 1
  for i in range(4):
      denominator = planet[i] + earth_values[i]
      if denominator == 0:
        continue

      diff = abs((planet[i] - earth_values[i]) / denominator)
      bracket = 1 - diff

      result = bracket ** (earth_weights[i] / 4)
      tracker_product *= result

  if tracker_product > esi:
    esi = tracker_product
    closest_planet = name

print("the closet body is: {} with an ESI of {}".format(closest_planet, esi))

```

```python
#pizza slices
def lazy_caterer_sequence(limit):
  """generates a list of the lazy caterer's sequence up to n = limit"""
  sequence = [1] # f(0) = 1
  for n in range (1, limit + 1):
    #f(n) = f(n-1) + n
    next_value = sequence[-1] + n
    sequence.append(next_value)
  return sequence

print(lazy_caterer_sequence(10))
```

```python
def question(a, b):
    """old school addition"""

    # 1. Convert the numeric inputs into strings.
    # We do this because numeric types don't have lengths, but we need
    # to know how many digits they have to calculate the layout.
    str_a = str(a)
    str_b = str(b)

    # 2. Calculate the total width of the text block.
    # max() finds whichever number has more digits.
    # We add 2 to that length to leave room for the '+' sign and a space.
    width = max(len(str_a), len(str_b)) + 2

    # 3. Print the top number (a).
    # .rjust(width) pads the left side with spaces so the entire
    # string matches the total 'width', pushing the digits to the right.
    print(str_a.rjust(width))

    # 4. Print the bottom number (b) with the plus sign.
    # Since the '+' sign takes up 1 character slot on the far left,
    # we only right-justify the number itself to (width - 1).
    print(f"+{str_b.rjust(width - 1)}")

    # 5. Print the mathematical separator line.
    # This multiplies the '-' character by our total calculated width.
    print('-' * width)

    # 6. Calculate the sum, convert it to a string, and print it.
    # Just like the top number, it is right-justified to the total width
    # so the ones, tens, and hundreds columns align perfectly.
    print(str(a + b).rjust(width))
```

This is absolutely not cheating! Breaking down a problem into logical steps before writing a single line of code is exactly how great programmers think.

Let’s dismantle this Scrabble problem step-by-step so you can map out the logic.

Step 1: Understand the Grid & Coordinate System
Before writing code, we need to understand how the physical Scrabble board translates to computer coordinates.

The board is a 15x15 grid.

Rows are labeled A to O.

Columns are labeled 1 to 15.

In programming, we usually prefer working with 0-based index numbers (from 0 to 14) for both rows and columns.

Think about: > * How will you convert the letter 'A' to index 0, 'B' to 1, and 'O' to 14? (Hint: Characters have underlying numeric ASCII/Unicode values).

How will you convert the column string '1' to index 0, and '15' to 14?

Step 2: Parse the Input String
Your function will receive a starting position like 'G7' or 'K12'.
You need to split this string into two parts: the row letter and the column number.

The Trap: You can't just split the string exactly in half because the column might be a single digit (like '7') or a double digit (like '12').

The Solution: The row is always the very first character. The column is everything that follows it.

Once you separate them, convert them to your 0-based grid indices from Step 1. Let’s call these start_row and start_col.

Step 3: Calculate the "Footprint" of the Word
A word has a certain length (let's call it L).
Depending on the direction ('across' or 'down'), the word will stretch out from its starting point.

If the word goes "across" (horizontally):

The row index stays the same.

The column index increases. If it starts at start_col, what index will the last letter of the word land on?

If the word goes "down" (vertically):

The column index stays the same.

The row index increases. If it starts at start_row, what index will the last letter of the word land on?

Mathematical Check: If a word of length 4 starts at index 12, what are the indices of the four letters? (They would be 12, 13, 14, and 15).

Step 4: Perform the Boundary Check
Now that you know where the word ends, you just need to make sure it doesn't "fall off" the edge of the board.

Since our grid indices only go from 0 to 14:

The starting index must be valid (not less than 0).

The ending index must not exceed 14.

If the ending index is within the boundaries, the word fits! Otherwise, it doesn't.

Which of these four steps feels like the trickiest one to implement in your preferred programming language?

```python

#--INPUT--
position = 'C9' # letters are columns and rows are numbers
word_a = 'scrabble'
direction = 'across'


#--FUNCTION DEFINITION --
def grid_system(position, word_a, direction):
  """takes a starting position, word and direction and returns tures if the word fits on a standard scrabble board"""
  coordinates = [] # initialise
  column_start = str(position[0]) # extract
  column_start = column_start.lower() # normalise
  column_start = ord(column_start) - 96 # convert to ascii --> alphabet position
  row_start = int(position[1:]) # extract everything after the first character and turn into int
  coordinates.append(column_start)
  coordinates.append(row_start)

  #work out where
  L = len(word_a) # determines how many  tiles it will occupy
  column_end = coordinates[0] + L - 1 # calculates the ending column index, and subtracts 1 as it is occupying the first tile
  row_end = coordinates[1] + L - 1 # calculates the ending row, and substracts 1 as it is occupying the first tile

#-- BOUNDARY CHECKING --
  if direction == 'across': #if placed horizontally check if ending column is within the boundary
    return column_end <= 15
  elif direction == 'down': #if placed vertically, checking boundaries
    return row_end <= 15
  else:
    return False

result = grid_system(position, word_a, direction)
print(result)








```

```python
# Exemplar Scrabble solution..
def word_fits(word, position, across):
  """determine if a give word fits in a scrabble grid"""
  if across:
    # get the starting column position
    i = int(position[1:])
  else:
    # get teh starting row position
    i = ord(position[0]) - 64
  return len(word) <= 16 - i # determine the direction string for the final product

def report(word, position, across):
    across_down = 'across' if across else 'down'# determine the string for the final printout
    mod = 'not ' if not word_fits(word, position, across) else ''
    return '"{} {} at {} does {} fit "'.format(word, across_down, position, mod)

print(report('HELLO', 'I2', across=False))

```

```python
#2.7.2 number problem

def sum_digits(n):
  """return the sum of the decimal digits of n"""
  s = 0 #initialise a running total for the sum of the digits
  while n: # loops as long as n is greater than 0
  #1. n % 10 grabs the last digit (remainder when divided by 10)
  #2. add that digit to the running total
    s += n % 10
    # use floor division to strip last digit
    # eg 123 // 10 = 12 eventually n becomes 0 adn the loop ends
    n //= 10
  return s

#multiple assignment: initialises both 'n' (our counter)
#and nfac (running factorial prodcut) to 1
n = nfac = 1

while True: # starts an infinite loop; it will run until it hits 'break'
  #efficiently calculates next factorial
  #instead of recalculating from scratch, it multiplies the previous
  #factorial by the new 'n' (e.g. 3! * 4 = 4!)
  nfac *= n
  digit_sum = sum_digits(nfac) #calls our helper function to get the sum of the digits of the current factorial

  # the % modulo operator returns teh remainder nfac / digitsum
  # if there is a remainder, the result is a non zero interger
  # means if NOT divisible by
  if nfac % digit_sum:
    # condition is met
    print('{}! is not divisible by sum of the digits'.format(n))
    break
  n += 1 # make sure it is inside the while loop
```
