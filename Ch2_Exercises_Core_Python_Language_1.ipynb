```python
#Q2.28 How many times must a sheet of paper (thickness, t = 0.1mm but otherwise any size required)
#be folded to reach the Moon (distance from Earth, d = 384400 km)
import math as m
t = 0.1e-3
d = 3.84400e8
#Workings
# fold 1 = t*2
# 1 = d / (t*2**N) N is the total number of folds
# t*2**N = d
# 2**N = d / t
# N = log2(d/t)
N = m.log2(d/t)
print(m.ceil(N))

#Exemplar solution
t = 1.e-4 # paper thickness, m
d = 384_400 * 1.e3 # distance to Moon, m
m.ceil(m.log(d/t, 2)) # base-2 logarituhm
```

```python
#P2.4.7 (a) Benford's law applied to Fibonacci sequence n = 500

#calculate ans stores the first n Fibonacci numbers
n = 500
fib = [0,1]
fib_digits_only = []
for i in range(2,n+1):
  next_fib = fib[i-1] + fib[i-2]
  fib.append(next_fib)
  fib_digits_only.append(int(str(next_fib)[0])) #extract first digits

ones_count = fib_digits_only.count(1)
twos_count = fib_digits_only.count(2)
threes_count = fib_digits_only.count(3)
fours_count = fib_digits_only.count(4)
fives_count = fib_digits_only.count(5)
sixes_count = fib_digits_only.count(6)
sevens_count = fib_digits_only.count(7)
eights_count = fib_digits_only.count(8)
nines_count = fib_digits_only.count(9)

ones_distribution = ones_count / len(fib_digits_only)
twos_distribution = twos_count / len(fib_digits_only)
threes_distribution = threes_count / len(fib_digits_only)
fours_distribution = fours_count / len(fib_digits_only)
fives_distribution = fives_count / len(fib_digits_only)
sixes_distribution = sixes_count / len(fib_digits_only)
sevens_distribution = sevens_count / len(fib_digits_only)
eights_distribution = eights_count / len(fib_digits_only)
nines_distribution = nines_count / len(fib_digits_only)

print(round(ones_distribution,3),
      round(twos_distribution,3),
      round(threes_distribution,3),
      round(fours_distribution,3),
      round(fives_distribution,3),
      round(sixes_distribution,3),
      round(sevens_distribution,3),
      round(eights_distribution,3),
      round(nines_distribution,3)
      )
```

```python
#P2.4.5
dna = 'AGTCFTATATCT'

# A dictionary to hold the codons for all three frames
reading_frames = {1: [], 2: [], 3: []}

# Loop through starting positions: 0, 1, and 2
for frame in range(0, 3):
    for i in range(frame, len(dna), 3):
        codon = dna[i : i + 3]
        # Only add complete 3-letter codons
        if len(codon) == 3:
            reading_frames[frame + 1].append(codon)

print("Frame 1:", reading_frames[1])
print("Frame 2:", reading_frames[2])
print("Frame 3:", reading_frames[3])
```

```python
# P.2.4.6 (a)
import math  # 1. Import the math module to use factorial

n = range(10)

for i in n:
    # 2. Check if the loop variable 'i' is EVEN.
    # If it is even, we skip it because we only want odd numbers.
    if i % 2 == 0:
        continue

    # 3. If it wasn't even, it must be odd!
    else:
        odd = math.factorial(i) # 4. Use math.factorial()
        print(f"The factorial of odd number {i} is {odd}")
```

```python
# pascal's triangle = (x + y)^n

import sympy as sp
x, y = sp.symbols('x y')

for i in range(0,6):
    expression = (x + y)**i
    #print(expression)
    expanded_expression = sp.expand(expression)
    #print(expanded_expression)

    if i == 0:
      coefficients = [1]
    else:
      clean_expression = expanded_expression.subs(y,1)

      poly = sp.Poly(clean_expression,x)
      coefficients = poly.all_coeffs()
    print(f"Coefficients {i}: {coefficients}")


```
