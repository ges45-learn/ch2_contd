```python
#2.5.1 exemplar, pascal's triangle using a nested list

# INITIALISTION
nmax = 4 # 5 rows
# Initialize with the first two rows
rows = [[1], [1, 1]] # pre-loaded with first rows

# OUTERLOOP (ROW BY ROW)
for n in range(2, nmax + 1):
    rows.append([1]) # new row by placing 1 at the beginning

    prev_row = rows[n - 1] # gets the completed row right above so we can look at the numbers

# INNER LOOP (MATH & COLUMN CREATION), WHERE THE MAGIC HAPPENS

    for k in range(len(prev_row) - 1):
        rows[n].append(prev_row[k] + prev_row[k + 1])

    rows[n].append(1) # End the row with a 1

#'Pretty-print the rows as asterisks for odd numbers or else spaces'


for row in rows:
    s_row = []
    for x in row:
        if x % 2: # if number is odd x % 2 evaluates to --> True, so it appends *
            s_row.append('*')
        else:
            s_row.append(' ') # if even --> False
    # Optional: Center the output slightly for a better visual triangle
    print(''.join(s_row))
```

```python
#p2.5.5 determine tomorrow's date

# text extraction phase

today = '22/05/2024'

us_style_today = 0
uk_style_today = 1

# Split once up front to keep the code clean
date_parts = today.split('/')

if us_style_today == True:
    day = int(date_parts[1])
    month = int(date_parts[0])
    year = int(date_parts[2])
else:
    day = int(date_parts[0])
    month = int(date_parts[1])
    year = int(date_parts[2])

group_1 = [1, 3, 5, 7, 8, 10, 12]
group_2 = [4, 6, 9, 11]
group_3 = [2]

# CALCULATE LEAP YEAR UP FRONT: Moved here so it doesn't break the if/elif chain below
is_leap_year = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Main date rollover logic
if month in group_1:
    if day == 31:
        if month == 12:  # Safe catch for New Year's Eve
            month = 1
            year += 1
        else:
            month += 1
        day = 1
    else:
        day += 1

elif month in group_2:
    if day == 30:
        month += 1
        day = 1
    else:
        day += 1

elif month in group_3:  # Now this safely connects back to the main chain
    # Condition A: It's Leap Day (Feb 29) OR it's Feb 28 in a normal year
    if (is_leap_year and day == 29) or (not is_leap_year and day == 28):
        month = 3
        day = 1

    # Condition B: It's Feb 28 in a leap year -> Advance to Leap Day (Feb 29)
    elif is_leap_year and day == 28:
        day += 1

    # Condition C: Any other day in February (e.g., Feb 1st to 27th) -> Just add 1
    else:
        day += 1

print(f"Tomorrow is: {day:02d}/{month:02d}/{year}")

```

```python
def elec_config(Z=6):
  """
  Write a program to predict the electronic configurations of the elements up to
  Rutherfordium (N=104). The output for Titanium should be

  Ti: 1s2.2s2.2p6.3s2.3p6.4s2.3d2

  This function returns the electronic configuration of an atom with atomic
  number Z.
  Z is valid until Z = 104.

  Input:
  ------
  Z: int


  Output:
  ------
  config: str
  """
  # import element symbols
  import element_symbols as es
  symbols = es.element_symbols

  # Find the symbol associated with number Z
  atom_name = symbols[Z-1]
  elec_config = ""

  allocated = 0

  while allocated < Z:
    if allocated < 2:
      if elec_config[:2] != '1s':
        elec_config += '1s'
        allocated += 1
      else:
        elec_config += '2'
        allocated += 1


  config = atom_name + ": " + elec_config
  return config
```

```python
#!git clone https://github.com/ges45-learn/ch2_contd.git #! = run on terminal, not python code. clone the repo (can run in its own cell if needed)


!rm -rf ch2_contd
!git clone https://github.com/ges45-learn/ch2_contd.git

f = open('ch2_contd/redwood-data.txt','r')
tree_name, location, diameter, height = [], [], [], []

line_count = 0

for line in f:
  line_count += 1

  if line_count <= 2:
    continue

  line = line.strip()
  if not line:
    continue

  fields = line.split('\t') #split by tab

  #append to lists
  tree_name.append(fields[0])
  location.append(fields[1])
  diameter.append(float(fields[2]))
  height.append(float(fields[3]))


f.close()

# Create a dictionary where tree_name is the key
# and a dictionary of its attributes (location, diameter, height) is the value
trees_data_dict = {}
for i in range(len(tree_name)):
    trees_data_dict[tree_name[i]] = {
        'location': location[i],
        'diameter': diameter[i],
        'height': height[i]
    }

print(trees_data_dict)
```
